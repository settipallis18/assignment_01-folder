## Data Cleaning & Visualization

### Removed unwanted column

```
#Dropping unwanted columns

df_new = df.drop(columns=['Unnamed: 0'])
df_new.head()
```

### Displaying the unique values in the Indicator 

```
df.indicator.unique()
```

### Displaying the unique Countries list 

```
df['Country Name'].unique()
```

### Printing the values of particular year (2022)

```
#Taking only particular data of 2020 year

df_2020 = df_new.query("Year == 2020")
df_2020.sample(10)
```

### Finding the CO2 dsamage of particular year 2022 based on the data

```
#Filtering only Carbon Dioxide Damage

df_2020_co2 = df_2020.query("indicator in['NY.ADJ.DCO2.GN.ZS']")
df_2020_co2
```

### Plotted Bar graph based on Co2 damage data in the year 2020

```
#Plotting bar graph of Co2 damage(%GNI)


fig = px.bar(
    data_frame=df_2020_co2,
    title="Total Co2 Damage (GNI%) of the year 2020 ",
    labels={"value":"Co2 Damage (GNI%)"},
    x="Country Name",
    y="value",
    color="Country Name"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

### Finding the Population Growth dsamage of particular year 2020 based on the data

```
#Total Population Growth (annual%) in the year 2020

df_2020_Pop = df_2020.query("indicator in['SP.POP.GROW']")
df_2020_Pop
```

### Plotted Bar graph based on Population Growth data in the year 2020

```
#Plotting the bar graph for the year 2020

fig = px.bar(
    data_frame=df_2020_Pop,
    title="Population Growth (annual%) in the year 2020",
    labels={"value":"Population growth (annual%)", "Year":"YEAR"},
    x="Country Name",
    y="value",
    color="Country Code"
)

fig.update_xaxes(categoryorder="total descending")


fig.show()
```

### Pie graph for Total Population Growth (annual%) year **2020**

```
#Plotting pie graph for Total Population Growth (annual%) in the year 2020

fig = px.pie(
    df_2020_Pop,
    values='value', 
    names='Country Name',
    title='Population growth (annual%) in the year 2020',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

### pie graph for Total C02 damage(GNI%) year 2020

```
fig = px.pie(
    df_2020_co2,
    values='value', 
    names='Country Name',
    title='C02 damage(GNI%) in the year 2020',
    hover_name='Country Code',
)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

### Finding the Co2 Damage and Population Growth for particular country 

```
df_ind = df_new[df_new['Country Code']== 'IND']
df_ind
```

### CO2 Damage (GNI%) of India

```
# Co2 of India across the years
df_ind_env = df_ind.query("indicator in(['NY.ADJ.DCO2.GN.ZS','SP.POP.GROW'])")
df_ind_env
```

```
df_pivot = df_ind_env.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot
```

```
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "NY.ADJ.DCO2.GN.ZS",
    title = "Co2 Damage",
    labels = {"Year":"Year","NY.ADJ.DCO2.GN.ZS":"Co2"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    
)

fig.show()
```

```
fig = px.bar(
    df_pivot.dropna(),
    x = "Year",
    y = "SP.POP.GROW",
    title = "Population Growth",
    labels = {"Year":"Year","SP.POP.GROW":"Population"},
    color = "Region",
    hover_name = "Country Name",
    text = "Country Name",
    
)

fig.show()
```

### Plotting line All regions

```
df_pivot_cont1 = df.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'], columns ='indicator', values ='value').reset_index()
df_pivot_cont1
```

### CO2 Damage of countries in a line 

```
fig = px.line(
    df_pivot_cont1, 
    x="Year", 
    y="NY.ADJ.DCO2.GN.ZS",
    labels={"NY.ADJ.DCO2.GN.ZS":" Over all Co2 Damage (GNI%)"},
    hover_name='Country Name',
    color='Country Name'
    )

fig.show()
```

### Population Growth of countries in a line

```
fig = px.line(
    df_pivot_cont1, 
    x="Year", 
    y="SP.POP.GROW",
    labels={"SP.POP.GROW":" Over all Population Growth(annula%"},
    hover_name='Country Name',
    color='Country Name'
    )

fig.show()
```

### Plotting Scatter Plot for all countries 

```
fig = px.scatter(
    df_pivot_cont1.dropna(),
    x="NY.ADJ.DCO2.GN.ZS",
    y="SP.POP.GROW",
    title="TOP Countries with hig Co2 damage in 2016",
    labels={"NY.ADJ.DCO2.GN.ZS":"s) Total Co2 damage in the year 2016","SP.POP.GROW":"Total Population growth (annual%) in the year 2016"},
    color='Country Name',
    hover_name='Country Name',
    
    size_max=75,
    height =800
)
fig.update_layout(showlegend=False)
fig.update_traces(textposition="top center")
fig.show()
```
