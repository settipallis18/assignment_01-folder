## 3.Find out the definition of the above 9 variables and document the definitions in the Jupyter Notebook in a Markdown cell so that the readers of your notebook will know the meaning of these variables.
#### UNITID  : Data files are provided at the UNITID level, which is the unique identification number assigned to postsecondary institutions as surveyed through IPEDS. 
#### INSTNM:  The institution’s name (INSTNM), as reported in IPEDS
#### CITY:  The institution’s location, as reported in IPEDS (CITY, STABBR, ZIP). This is also reported using latitude and longitude (float; LATITUDE and LONGITUDE)
#### STABBR: The institution’s state code, as reported in IPEDS 
#### ZIP: This is the zip code of the institution's location (City's zipcode).
#### ADM_RATE:  Colleges report to IPEDS their Fall admissions rate, defined as the number of admitted undergraduates divided by the number of undergraduates who applied. For institutions with multiple branches, ADM_RATE includes the admissions rate at each campus
#### UGDS: This element (UGDS) includes the number of degree/certificate-seeking undergraduates enrolled in the fall, as reported in the IPEDS Fall Enrollment component.
#### TUITIONFEE_IN: the cost data include the tuition and required fees, booksand supplies, and living expenses estimated by the institution. Tuition and required fees are provided for in-state students (TUITIONFEE_IN)
#### MN_EARN_WNE_P10: Mean (MN_EARN_WNE_P*) earnings are for the institutional aggregate of all federally aided students who enroll in an institution each year and who are employed but not enrolled

#### Source: https://collegescorecard.ed.gov/assets/InstitutionDataDocumentation.pdf

```python
U1 = df1.loc[df1["INSTNM"].str.contains('University of Maryland-Baltimore County')]
U1
```

### e) Find Univ. of Maryland College Park

```python
U2 = df1.loc[df1["INSTNM"].str.contains('University of Maryland-College Park')]
U2
```

```python
U3 = df1.loc[df1["INSTNM"].str.contains('Loyola University Maryland')]
U3
```

### g) Find and display Johns Hopskins

```python
U4 = df1.loc[df1["INSTNM"].str.contains('Johns Hopkins University')]
U4
```

### h) Calculate the min, max, and average in-state tution of the above four institutions.

```python
# created a new datafrme for finding min, max & avg values by concatinating the all university values.
df2 = pd.concat([U1, U2, U3, U4], axis=0)
df2
```

```python
# finding the minimin in-state tuition fee of instituttions
min = df2['TUITIONFEE_IN'].min()
print("Minimum in-state tuition fee is", min)
```

```python
# finding the maximum in-state tuition fee of instituttions
max = df2['TUITIONFEE_IN'].max()
print("Maximum in-state tuition fee is", max)
```

```python
# finding the average in-state tuition fee of instituttions
avg = df2['TUITIONFEE_IN'].mean()
print("Average in-state tuition fee is", avg)
```
